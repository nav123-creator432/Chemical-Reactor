```python
import numpy as np
import pandas as pd
from scipy.integrate import odeint
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import TimeSeriesSplit
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.ensemble import RandomForestRegressor
import xgboost as xgb
import lightgbm as lgb
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import optuna
import shap
optuna.logging.set_verbosity(optuna.logging.WARNING)

# ─────────────────────────────────────────────
# CONSTANTS
# ─────────────────────────────────────────────
R = 8.314
Ea1, Ea2, Ea3 = 50000, 70000, 60000   # J/mol — physically realistic, temperature-sensitive


A1 = 5.1e7    # k1(300K) ≈ 0.08  →  A peaks around t=8h
A2 = 2.0e6    # k2(300K) ≈ 0.003 →  B→C is slow, giving B a nice peak then decay
A3 = 1.5e7    # k3(300K) ≈ 0.023 →  side reaction, smaller than main


# ─────────────────────────────────────────────
# IDEAL MODEL (constant T=300K, perfect catalyst)
# ─────────────────────────────────────────────
def ideal_kinetics(y, t):
    A, B, C, D = y
    T = 300
    k1 = A1 * np.exp(-Ea1 / (R * T))
    k2 = A2 * np.exp(-Ea2 / (R * T))
    k3 = A3 * np.exp(-Ea3 / (R * T))
    return [-(k1 + k3) * A,
             k1 * A - k2 * B,
             k2 * B,
             k3 * A]

# ─────────────────────────────────────────────
# REAL INDUSTRIAL MODEL
# ─────────────────────────────────────────────
def real_system(y, t):
    A, B, C, D, Activity = y
    T = 300 + 25 * np.sin(2 * np.pi * t / 24)   # ±25K day-night swing
    k1 = A1 * np.exp(-Ea1 / (R * T)) * Activity
    k2 = A2 * np.exp(-Ea2 / (R * T)) * Activity
    k3 = A3 * np.exp(-Ea3 / (R * T)) * Activity
    k_deact = 0.05                                # catalyst deactivation
    return [-(k1 + k3) * A,
             k1 * A - k2 * B,
             k2 * B,
             k3 * A,
            -k_deact * Activity * A]

# ─────────────────────────────────────────────
# GENERATE DATA
# ─────────────────────────────────────────────
np.random.seed(42)
t = np.linspace(0, 48, 1000)

ideal_sol = odeint(ideal_kinetics, [1, 0, 0, 0],    t)
real_sol  = odeint(real_system,    [1, 0, 0, 0, 1], t)

df = pd.DataFrame({
    'time':     t,
    'ideal_B':  ideal_sol[:, 1],
    'real_B':   real_sol[:,  1],
    'temp':     300 + 25 * np.sin(2 * np.pi * t / 24),  # matches real_system
    'activity': real_sol[:, 4],
})

# ─────────────────────────────────────────────
# SENSOR NOISE + MISSING DATA
# ─────────────────────────────────────────────
df['real_B'] += np.random.normal(0, 0.002, size=len(df))
df.loc[df.sample(frac=0.05, random_state=0).index, 'real_B'] = np.nan

# valid mask BEFORE interpolation — marks originally real (non-NaN) rows
valid_mask = ~df['real_B'].isna()

# ─────────────────────────────────────────────
# PREPROCESSING — interpolation instead of median (no spike contamination)
# ─────────────────────────────────────────────
df['real_B_imputed'] = df['real_B'].interpolate(method='linear').bfill()

df['B_lag1']  = df['real_B_imputed'].shift(1).bfill()
df['B_lag2']  = df['real_B_imputed'].shift(2).bfill()
df['temp_sq'] = df['temp'] ** 2
df['residual'] = df['real_B_imputed'] - df['ideal_B']

features = ['time', 'temp', 'temp_sq', 'ideal_B', 'B_lag1', 'B_lag2', 'activity']
X = df[features].values
y = df['residual'].values

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ─────────────────────────────────────────────
# TRAIN / TEST SPLIT — valid rows only
# ─────────────────────────────────────────────
X_valid     = X_scaled[valid_mask]
y_valid     = y[valid_mask]
t_valid_idx = np.where(valid_mask)[0]

split    = int(len(X_valid) * 0.80)
X_train  = X_valid[:split];   y_train = y_valid[:split]
X_test   = X_valid[split:];   y_test  = y_valid[split:]

X_train_df = pd.DataFrame(X_train, columns=features)
X_test_df  = pd.DataFrame(X_test,  columns=features)
X_all_df   = pd.DataFrame(X_scaled, columns=features)

# ─────────────────────────────────────────────
# OPTUNA HYPERPARAMETER TUNING
# ─────────────────────────────────────────────
def xgb_objective(trial):
    params = {
        'n_estimators':     trial.suggest_int('n_estimators',   100, 600),
        'max_depth':        trial.suggest_int('max_depth',        3,   8),
        'learning_rate':    trial.suggest_float('learning_rate', 0.01, 0.3, log=True),
        'subsample':        trial.suggest_float('subsample',      0.6,  1.0),
        'colsample_bytree': trial.suggest_float('colsample_bytree', 0.6, 1.0),
        'random_state': 42,
        'verbosity': 0,
    }
    tscv = TimeSeriesSplit(n_splits=5)
    maes = []
    for tr_idx, val_idx in tscv.split(X_train):
        m = xgb.XGBRegressor(**params)
        m.fit(X_train[tr_idx], y_train[tr_idx])
        maes.append(mean_absolute_error(y_train[val_idx], m.predict(X_train[val_idx])))
    return np.mean(maes)

study = optuna.create_study(direction='minimize')
study.optimize(xgb_objective, n_trials=50)
best_params = study.best_params
print("Best XGB params:", best_params)

# ─────────────────────────────────────────────
# TRAIN ALL THREE MODELS
# ─────────────────────────────────────────────
xgb_model = xgb.XGBRegressor(**best_params, random_state=42, verbosity=0)
xgb_model.fit(X_train, y_train)

lgb_model = lgb.LGBMRegressor(
    n_estimators=400, learning_rate=0.05,
    max_depth=5, random_state=42, verbose=-1)
lgb_model.fit(X_train_df, y_train)

rf_model = RandomForestRegressor(n_estimators=300, max_depth=8, random_state=42)
rf_model.fit(X_train, y_train)

# ─────────────────────────────────────────────
# HYBRID PREDICTIONS
# ─────────────────────────────────────────────
df['hybrid_xgb'] = df['ideal_B'] + xgb_model.predict(X_scaled)
df['hybrid_lgb'] = df['ideal_B'] + lgb_model.predict(X_all_df)
df['hybrid_rf']  = df['ideal_B'] + rf_model.predict(X_scaled)

ideal_test       = df['ideal_B'].values[t_valid_idx][split:]
real_test        = df['real_B_imputed'].values[t_valid_idx][split:]
hybrid_xgb_test  = ideal_test + xgb_model.predict(X_test)
hybrid_lgb_test  = ideal_test + lgb_model.predict(X_test_df)
hybrid_rf_test   = ideal_test + rf_model.predict(X_test)

# ─────────────────────────────────────────────
# EVALUATION
# ─────────────────────────────────────────────
# EVALUATION — active region only
# ─────────────────────────────────────────────
def evaluate(name, y_true, y_pred):
    mae  = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    r2   = r2_score(y_true, y_pred)
    print(f"{name:22s} | MAE={mae:.5f}  RMSE={rmse:.5f}  R²={r2:.5f}")
    return mae, rmse, r2

# Active region mask — where B actually matters
active_mask = df['ideal_B'] > 0.01

# Get active region values for each model
real_active    = df.loc[active_mask, 'real_B_imputed'].values
ideal_active   = df.loc[active_mask, 'ideal_B'].values
xgb_active     = df.loc[active_mask, 'hybrid_xgb'].values
lgb_active     = df.loc[active_mask, 'hybrid_lgb'].values
rf_active      = df.loc[active_mask, 'hybrid_rf'].values

print("\n--- MODEL PERFORMANCE (active region: B > 0.01) ---")
evaluate("XGBoost Hybrid",      real_active, xgb_active)
evaluate("LightGBM Hybrid",     real_active, lgb_active)
evaluate("RandomForest Hybrid", real_active, rf_active)
evaluate("Ideal Model Only",    real_active, ideal_active)

print("\nNote: Metrics computed on active region only (where B > 0.01)")
print("This is the physically meaningful window for reactor decision-making.")



# ─────────────────────────────────────────────
# FEATURE IMPORTANCE — SHAP
# ─────────────────────────────────────────────
explainer   = shap.Explainer(xgb_model, feature_names=features)
shap_values = explainer(pd.DataFrame(X_test, columns=features))

# ─────────────────────────────────────────────
# PROFIT OPTIMIZATION
# ─────────────────────────────────────────────
def profit(yield_b, t_arr, price=1000, op_cost=20, catalyst_cost=50):
    return (yield_b * price) - (t_arr * op_cost) - catalyst_cost

df['profit_hybrid'] = profit(df['hybrid_xgb'], df['time'])
df['profit_ideal']  = profit(df['ideal_B'],    df['time'])

opt_time   = df.loc[df['profit_hybrid'].idxmax(), 'time']
ideal_time = df.loc[df['profit_ideal'].idxmax(),  'time']

print("\n--- DECISION REPORT ---")
print(f"Ideal model stop time : {ideal_time:.2f} h")
print(f"Hybrid model stop time: {opt_time:.2f} h")
print(f"Improvement           : {abs(opt_time - ideal_time):.2f} h")

# ─────────────────────────────────────────────
# PLOTS
# ─────────────────────────────────────────────
fig = plt.figure(figsize=(16, 14))
gs  = gridspec.GridSpec(3, 2, figure=fig, hspace=0.42, wspace=0.32)

# 1. Concentration profiles
ax1 = fig.add_subplot(gs[0, :])
ax1.plot(t, df['ideal_B'],    'b--', lw=2,   label='Ideal model')
ax1.plot(t, df['hybrid_xgb'], 'r-',  lw=2,   label='Hybrid (XGBoost)')
ax1.plot(t, df['real_B'],     'k-',  lw=1.2, label='Real (industrial)', zorder=5)
# black plotted LAST so it sits on top of red
imputed_idx = np.where(~valid_mask)[0]
ax1.scatter(t[imputed_idx], df.loc[imputed_idx, 'real_B_imputed'],
            color='orange', s=14, zorder=6, label='Imputed points', alpha=0.8)
ax1.set_xlabel('Time (h)')
ax1.set_ylabel('Concentration of B')
ax1.set_title('Concentration of product B over time')
ax1.legend()
ax1.grid(alpha=0.3)

# 2. Parity plot — non-imputed test points only
ax2 = fig.add_subplot(gs[1, 0])
active_mask = df['ideal_B'] > 0.01
real_act = df.loc[active_mask, 'real_B_imputed'].values
xgb_act  = df.loc[active_mask, 'hybrid_xgb'].values
ax2.scatter(real_act, xgb_act, alpha=0.3, s=8, color='steelblue')
lim = [min(real_act.min(), xgb_act.min()), max(real_act.max(), xgb_act.max())]
ax2.plot(lim, lim, 'r--', lw=1)
ax2.set_xlabel('Actual'); ax2.set_ylabel('Predicted')
ax2.set_title('Parity plot (active region, B > 0.01)')
ax2.grid(alpha=0.3)
# 3. Residuals vs time
ax3 = fig.add_subplot(gs[1, 1])
test_times = df['time'].values[t_valid_idx][split:]
ax3.scatter(test_times, real_test - hybrid_xgb_test,
            alpha=0.5, s=10, color='coral')
ax3.axhline(0, color='black', lw=1)
ax3.set_xlabel('Time (h)')
ax3.set_ylabel('Residual')
ax3.set_title('Residuals vs time (test set)')
ax3.grid(alpha=0.3)

# 4. SHAP feature importance
ax4 = fig.add_subplot(gs[2, 0])
shap.plots.bar(shap_values, max_display=7, show=False, ax=ax4)
ax4.set_title('SHAP feature importance (XGBoost)')

# 5. Profit optimization
ax5 = fig.add_subplot(gs[2, 1])
ax5.plot(t, df['profit_hybrid'], 'r-',  lw=1.5, label='Hybrid model')
ax5.plot(t, df['profit_ideal'],  'b--', lw=1.5, label='Ideal model')
ax5.axvline(opt_time,   color='red',  ls=':', lw=1.2,
            label=f'Hybrid stop: {opt_time:.2f}h')
ax5.axvline(ideal_time, color='blue', ls=':', lw=1.2,
            label=f'Ideal stop:  {ideal_time:.2f}h')
ax5.set_xlabel('Time (h)')
ax5.set_ylabel('Profit')
ax5.set_title('Profit optimization')
ax5.legend(fontsize=9)
ax5.grid(alpha=0.3)

plt.suptitle('Hybrid Physics-ML Model — Industrial Reactor', fontsize=14, y=1.01)
plt.savefig('hybrid_model_results.png', dpi=150, bbox_inches='tight')
plt.show()
print("\nPlot saved: hybrid_model_results.png")
```
