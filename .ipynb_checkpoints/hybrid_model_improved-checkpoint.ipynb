# Improved Hybrid Physics-ML Model — Industrial Reactor

### Improvements over baseline:
1. **Savitzky-Golay smoothing** on lag features → eliminates concentration spikes
2. **Richer features**: rolling mean/std, explicit sinusoidal temperature periodicity
3. **Optuna tuning for all 3 models** (XGBoost + LightGBM + RandomForest)
4. **Physical clipping** of predictions to valid concentration range [0, 1]
5. **Weighted ensemble model** (inverse-MAE weights)
6. **Uncertainty quantification** from RF tree variance (±2σ band)
7. **Residual trend line** to detect systematic drift
8. **Profit gain shading** on the optimization plot


```python
import numpy as np
import pandas as pd
from scipy.integrate import odeint
from scipy.signal import savgol_filter
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.model_selection import TimeSeriesSplit
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.ensemble import RandomForestRegressor
import xgboost as xgb
import lightgbm as lgb
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import optuna
import shap
import warnings
warnings.filterwarnings('ignore')
optuna.logging.set_verbosity(optuna.logging.WARNING)
```

```python
# ─────────────────────────────────────────────
# CONSTANTS  (Arrhenius — illustrative values)
```

```python
# ─────────────────────────────────────────────
R = 8.314
Ea1, Ea2, Ea3 = 5000, 7000, 6000
A1,  A2,  A3  = 1.0,  0.8,  0.5
```

```python
# ─────────────────────────────────────────────
# IDEAL MODEL  (constant T, no deactivation)
```

```python
# ─────────────────────────────────────────────
def ideal_kinetics(y, t):
    A, B, C, D = y
    T = 300
    k1 = A1 * np.exp(-Ea1 / (R * T))
    k2 = A2 * np.exp(-Ea2 / (R * T))
    k3 = A3 * np.exp(-Ea3 / (R * T))
    return [-(k1+k3)*A,  k1*A - k2*B,  k2*B,  k3*A]
```

```python
# ─────────────────────────────────────────────
# REAL INDUSTRIAL MODEL
```

```python
# ─────────────────────────────────────────────
def real_system(y, t):
    A, B, C, D, Activity = y
    T = 300 + 10 * np.sin(2 * np.pi * t / 24)
    k1 = A1 * np.exp(-Ea1 / (R * T)) * Activity
    k2 = A2 * np.exp(-Ea2 / (R * T)) * Activity
    k3 = A3 * np.exp(-Ea3 / (R * T)) * Activity
    k_deact = 0.01
    return [-(k1+k3)*A, k1*A - k2*B, k2*B, k3*A, -k_deact * Activity * A]
```

```python
# ─────────────────────────────────────────────
# GENERATE DATA
```

```python
# ─────────────────────────────────────────────
np.random.seed(42)
t = np.linspace(0, 48, 1000)

ideal_sol = odeint(ideal_kinetics, [1, 0, 0, 0],    t)
real_sol  = odeint(real_system,    [1, 0, 0, 0, 1], t)

df = pd.DataFrame({
    'time':     t,
    'ideal_B':  ideal_sol[:, 1],
    'real_B':   real_sol[:,  1],
    'temp':     300 + 10 * np.sin(2 * np.pi * t / 24),
    'activity': real_sol[:, 4],
})
```

```python
# ─────────────────────────────────────────────
# ADD SENSOR NOISE
```

```python
# ─────────────────────────────────────────────
df['real_B'] += np.random.normal(0, 0.002, size=len(df))
```

```python
# ─────────────────────────────────────────────
# PREPROCESSING — IMPROVED
```

```python
# ─────────────────────────────────────────────
df.loc[df.sample(frac=0.05, random_state=0).index, 'real_B'] = np.nan

imputer = SimpleImputer(strategy='median')
df['real_B_imputed'] = imputer.fit_transform(df[['real_B']])

# IMPROVEMENT 1: Smooth the imputed signal BEFORE computing lags
# This prevents noise/imputation artifacts from propagating into lag features
df['real_B_smooth'] = savgol_filter(df['real_B_imputed'], window_length=11, polyorder=3)

# IMPROVEMENT 2: Richer, noise-resistant features
df['B_lag1']        = df['real_B_smooth'].shift(1).bfill()
df['B_lag2']        = df['real_B_smooth'].shift(2).bfill()
df['B_roll5_mean']  = df['real_B_smooth'].rolling(5, min_periods=1).mean()   # NEW
df['B_roll5_std']   = df['real_B_smooth'].rolling(5, min_periods=1).std().fillna(0)  # NEW — local volatility
df['temp_sq']       = df['temp'] ** 2
df['temp_sin']      = np.sin(2 * np.pi * df['time'] / 24)                    # NEW — explicit periodicity
df['residual']      = df['real_B_imputed'] - df['ideal_B']

features = ['time', 'temp', 'temp_sq', 'temp_sin',
            'ideal_B', 'B_lag1', 'B_lag2', 'B_roll5_mean', 'B_roll5_std', 'activity']

X = df[features].values
y = df['residual'].values

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# ─────────────────────────────────────────────
# TRAIN / TEST SPLIT  (last 20% = held-out test)
```

```python
# ─────────────────────────────────────────────
split = int(len(X_scaled) * 0.80)
X_train, X_test = X_scaled[:split], X_scaled[split:]
y_train, y_test = y[:split],        y[split:]

X_train_df = pd.DataFrame(X_train, columns=features)
X_test_df  = pd.DataFrame(X_test,  columns=features)
X_all_df   = pd.DataFrame(X_scaled, columns=features)
```

```python
# ─────────────────────────────────────────────
# IMPROVEMENT 3: Optuna tuning for ALL three models
```

```python
# ─────────────────────────────────────────────
def tune_model(objective_fn, n_trials=40):
    study = optuna.create_study(direction='minimize')
    study.optimize(objective_fn, n_trials=n_trials)
    return study.best_params

# ── XGBoost ──
def xgb_objective(trial):
    params = {
        'n_estimators':      trial.suggest_int('n_estimators',       100, 600),
        'max_depth':         trial.suggest_int('max_depth',            3,   8),
        'learning_rate':     trial.suggest_float('learning_rate',    0.01, 0.3, log=True),
        'subsample':         trial.suggest_float('subsample',         0.6,  1.0),
        'colsample_bytree':  trial.suggest_float('colsample_bytree',  0.6,  1.0),
        'min_child_weight':  trial.suggest_int('min_child_weight',     1,   10),  # NEW
        'reg_alpha':         trial.suggest_float('reg_alpha',         0.0,  1.0),  # NEW — L1
        'reg_lambda':        trial.suggest_float('reg_lambda',        0.5,  5.0),  # NEW — L2
        'random_state': 42, 'verbosity': 0,
    }
    tscv = TimeSeriesSplit(n_splits=5)
    return np.mean([
        mean_absolute_error(y_train[v], xgb.XGBRegressor(**params).fit(X_train[tr], y_train[tr]).predict(X_train[v]))
        for tr, v in tscv.split(X_train)
    ])

# ── LightGBM ── IMPROVEMENT 3a: now also Optuna-tuned
def lgb_objective(trial):
    params = {
        'n_estimators':     trial.suggest_int('n_estimators',       100, 600),
        'num_leaves':       trial.suggest_int('num_leaves',          20, 150),
        'learning_rate':    trial.suggest_float('learning_rate',    0.01, 0.3, log=True),
        'subsample':        trial.suggest_float('subsample',         0.6,  1.0),
        'colsample_bytree': trial.suggest_float('colsample_bytree',  0.6,  1.0),
        'min_child_samples':trial.suggest_int('min_child_samples',    5,  50),
        'reg_alpha':        trial.suggest_float('reg_alpha',         0.0,  1.0),
        'reg_lambda':       trial.suggest_float('reg_lambda',        0.0,  5.0),
        'random_state': 42, 'verbose': -1,
    }
    tscv = TimeSeriesSplit(n_splits=5)
    return np.mean([
        mean_absolute_error(
            y_train[v],
            lgb.LGBMRegressor(**params).fit(
                pd.DataFrame(X_train[tr], columns=features), y_train[tr]
            ).predict(pd.DataFrame(X_train[v], columns=features))
        )
        for tr, v in tscv.split(X_train)
    ])

# ── RandomForest ── IMPROVEMENT 3b: now also Optuna-tuned
def rf_objective(trial):
    params = {
        'n_estimators': trial.suggest_int('n_estimators', 100, 400),
        'max_depth':    trial.suggest_int('max_depth',      4,  15),
        'max_features': trial.suggest_float('max_features', 0.4, 1.0),
        'min_samples_split': trial.suggest_int('min_samples_split', 2, 10),
        'random_state': 42,
    }
    tscv = TimeSeriesSplit(n_splits=5)
    return np.mean([
        mean_absolute_error(y_train[v], RandomForestRegressor(**params).fit(X_train[tr], y_train[tr]).predict(X_train[v]))
        for tr, v in tscv.split(X_train)
    ])

print("Tuning XGBoost...")
best_xgb = tune_model(xgb_objective, n_trials=50)
print("Tuning LightGBM...")
best_lgb = tune_model(lgb_objective, n_trials=40)
print("Tuning RandomForest...")
best_rf  = tune_model(rf_objective,  n_trials=30)

print("Best XGB params:", best_xgb)
print("Best LGB params:", best_lgb)
print("Best RF  params:", best_rf)
```

```python
# ─────────────────────────────────────────────
# TRAIN FINAL MODELS WITH BEST PARAMS
```

```python
# ─────────────────────────────────────────────
xgb_model = xgb.XGBRegressor(**best_xgb, random_state=42, verbosity=0)
xgb_model.fit(X_train, y_train)

lgb_model = lgb.LGBMRegressor(**best_lgb, random_state=42, verbose=-1)
lgb_model.fit(X_train_df, y_train)

rf_model = RandomForestRegressor(**best_rf, random_state=42)
rf_model.fit(X_train, y_train)
```

```python
# ─────────────────────────────────────────────
# HYBRID PREDICTIONS + IMPROVEMENT 4: Physical clipping [0, 1]
```

```python
# ─────────────────────────────────────────────
ideal_all  = df['ideal_B'].values
ideal_test = ideal_all[split:]
real_test  = df['real_B_imputed'].values[split:]

# Raw hybrid predictions
raw_xgb_correction = xgb_model.predict(X_scaled)
raw_lgb_correction = lgb_model.predict(X_all_df)
raw_rf_correction  = rf_model.predict(X_scaled)

# IMPROVEMENT 4: Clip to valid concentration range [0, 1]
df['hybrid_xgb'] = np.clip(ideal_all + raw_xgb_correction, 0, 1)
df['hybrid_lgb'] = np.clip(ideal_all + raw_lgb_correction, 0, 1)
df['hybrid_rf']  = np.clip(ideal_all + raw_rf_correction,  0, 1)

# IMPROVEMENT 5: Smooth final predictions with Savitzky-Golay (preserves peaks)
df['hybrid_xgb'] = savgol_filter(df['hybrid_xgb'], window_length=15, polyorder=3)
df['hybrid_lgb'] = savgol_filter(df['hybrid_lgb'], window_length=15, polyorder=3)
df['hybrid_rf']  = savgol_filter(df['hybrid_rf'],  window_length=15, polyorder=3)

# IMPROVEMENT 6: Ensemble model (weighted by inverse validation MAE)
#   — gives more weight to better models automatically
val_maes = {
    'xgb': mean_absolute_error(y_train[-200:], xgb_model.predict(X_train[-200:])),
    'lgb': mean_absolute_error(y_train[-200:], lgb_model.predict(X_train_df.tail(200))),
    'rf':  mean_absolute_error(y_train[-200:], rf_model.predict(X_train[-200:])),
}
weights = {k: 1/v for k, v in val_maes.items()}
total_w = sum(weights.values())
w_xgb, w_lgb, w_rf = weights['xgb']/total_w, weights['lgb']/total_w, weights['rf']/total_w
print(f"\nEnsemble weights — XGB: {w_xgb:.3f}  LGB: {w_lgb:.3f}  RF: {w_rf:.3f}")

df['hybrid_ensemble'] = (
    w_xgb * df['hybrid_xgb'] +
    w_lgb * df['hybrid_lgb'] +
    w_rf  * df['hybrid_rf']
)

# Test-set slices
hybrid_xgb_test      = df['hybrid_xgb'].values[split:]
hybrid_lgb_test      = df['hybrid_lgb'].values[split:]
hybrid_rf_test       = df['hybrid_rf'].values[split:]
hybrid_ensemble_test = df['hybrid_ensemble'].values[split:]

# IMPROVEMENT 7: Uncertainty quantification from RF individual trees
rf_tree_preds_test = np.array([tree.predict(X_test) for tree in rf_model.estimators_])
rf_std_test        = rf_tree_preds_test.std(axis=0)
```

```python
# ─────────────────────────────────────────────
# EVALUATION
```

```python
# ─────────────────────────────────────────────
def evaluate(name, y_true, y_pred):
    mae  = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    r2   = r2_score(y_true, y_pred)
    print(f"{name:22s} | MAE={mae:.5f}  RMSE={rmse:.5f}  R²={r2:.5f}")
    return mae, rmse, r2

print("\n--- TEST SET METRICS ---")
evaluate("XGBoost Hybrid",    real_test, hybrid_xgb_test)
evaluate("LightGBM Hybrid",   real_test, hybrid_lgb_test)
evaluate("RandomForest Hybrid",real_test, hybrid_rf_test)
evaluate("Ensemble Hybrid",   real_test, hybrid_ensemble_test)
evaluate("Ideal Model Only",  real_test, ideal_test)

residual_variance = np.var(y_test)
print(f"\nTarget (residual) variance on test set: {residual_variance:.6f}")

# Active region metrics (where B is still reacting)
active_mask = df['ideal_B'] > 0.01
print(f"\nMetrics on active region (B > 0.01) only:")
for name, col in [("XGBoost", "hybrid_xgb"), ("LightGBM","hybrid_lgb"),
                  ("RF","hybrid_rf"), ("Ensemble","hybrid_ensemble")]:
    mae_a = mean_absolute_error(df.loc[active_mask,'real_B_imputed'], df.loc[active_mask, col])
    r2_a  = r2_score(df.loc[active_mask,'real_B_imputed'], df.loc[active_mask, col])
    print(f"  {name:12s} — MAE={mae_a:.5f}  R²={r2_a:.5f}")
```

```python
# ─────────────────────────────────────────────
# FEATURE IMPORTANCE  (SHAP)
```

```python
# ─────────────────────────────────────────────
explainer   = shap.Explainer(xgb_model, feature_names=features)
shap_values = explainer(pd.DataFrame(X_test, columns=features))
```

```python
# ─────────────────────────────────────────────
# PROFIT OPTIMIZATION — use best model (ensemble)
```

```python
# ─────────────────────────────────────────────
def profit(yield_b, t_arr, price=1000, op_cost=20, catalyst_cost=50):
    return (yield_b * price) - (t_arr * op_cost) - catalyst_cost

df['profit_ensemble'] = profit(df['hybrid_ensemble'], df['time'])
df['profit_ideal']    = profit(df['ideal_B'],         df['time'])

opt_time   = df.loc[df['profit_ensemble'].idxmax(), 'time']
ideal_time = df.loc[df['profit_ideal'].idxmax(),    'time']

print("\n--- DECISION REPORT ---")
print(f"Ideal model stop time    : {ideal_time:.2f} h")
print(f"Ensemble hybrid stop time: {opt_time:.2f} h")
print(f"Improvement              : {abs(opt_time - ideal_time):.2f} h earlier")
```

```python
# ─────────────────────────────────────────────
# PLOTS — IMPROVED LAYOUT
```

```python
# ─────────────────────────────────────────────
fig = plt.figure(figsize=(18, 16))
gs  = gridspec.GridSpec(3, 2, figure=fig, hspace=0.45, wspace=0.35)

# ── 1. Concentration profiles (now with uncertainty band) ──
ax1 = fig.add_subplot(gs[0, :])
ax1.plot(t, df['real_B_smooth'],    'k-',  lw=1.5, label='Real (industrial, smoothed)', zorder=5)
ax1.plot(t, df['ideal_B'],          'b--', lw=1.5, label='Ideal model', zorder=4)
ax1.plot(t, df['hybrid_ensemble'],  'r-',  lw=2,   label='Ensemble Hybrid', zorder=6)

# Uncertainty band from RF tree variance (extended to all indices)
rf_tree_preds_all = np.array([tree.predict(X_scaled) for tree in rf_model.estimators_])
rf_std_all        = rf_tree_preds_all.std(axis=0)
hybrid_ens_arr    = df['hybrid_ensemble'].values
ax1.fill_between(t,
                 np.clip(hybrid_ens_arr - 2*rf_std_all, 0, 1),
                 np.clip(hybrid_ens_arr + 2*rf_std_all, 0, 1),
                 color='red', alpha=0.12, label='±2σ uncertainty (RF)')

ax1.axvline(opt_time,   color='red',  ls=':', lw=1.5, label=f'Optimal stop: {opt_time:.1f}h')
ax1.axvline(ideal_time, color='blue', ls=':', lw=1.5, label=f'Ideal stop: {ideal_time:.1f}h')
ax1.set_xlabel('Time (h)'); ax1.set_ylabel('Conc. of B')
ax1.set_title('Concentration of Product B — Ensemble Hybrid vs Ideal', fontsize=12)
ax1.legend(loc='upper right', fontsize=8); ax1.grid(alpha=0.3)

# ── 2. Parity plot (ensemble) ──
ax2 = fig.add_subplot(gs[1, 0])
ax2.scatter(real_test, hybrid_ensemble_test, alpha=0.5, s=10, color='steelblue',
            label='Ensemble', zorder=5)
lim = [min(real_test.min(), hybrid_ensemble_test.min()),
       max(real_test.max(), hybrid_ensemble_test.max())]
ax2.plot(lim, lim, 'r--', lw=1.5, label='Perfect fit')
ax2.set_xlabel('Actual'); ax2.set_ylabel('Predicted')
ax2.set_title('Parity Plot — Ensemble (test set)', fontsize=11)
ax2.legend(fontsize=8); ax2.grid(alpha=0.3)

# ── 3. Residuals vs time + zero line + rolling mean trend ──
ax3 = fig.add_subplot(gs[1, 1])
resid = real_test - hybrid_ensemble_test
test_times = df['time'].values[split:]
ax3.scatter(test_times, resid, alpha=0.4, s=10, color='coral', label='Residuals')
ax3.axhline(0, color='black', lw=1)
# Add rolling mean of residuals to spot drift
resid_roll = pd.Series(resid).rolling(20, min_periods=1).mean().values
ax3.plot(test_times, resid_roll, 'navy', lw=1.5, label='Rolling mean (trend)')
ax3.set_xlabel('Time (h)'); ax3.set_ylabel('Residual')
ax3.set_title('Residuals vs Time — Ensemble (test set)', fontsize=11)
ax3.legend(fontsize=8); ax3.grid(alpha=0.3)

# ── 4. SHAP bar chart ──
ax4 = fig.add_subplot(gs[2, 0])
shap.plots.bar(shap_values, max_display=10, show=False, ax=ax4)
ax4.set_title('SHAP Feature Importance (XGBoost)', fontsize=11)

# ── 5. Profit curves ──
ax5 = fig.add_subplot(gs[2, 1])
ax5.plot(t, df['profit_ensemble'], 'r-',  lw=2,   label='Ensemble Hybrid')
ax5.plot(t, df['profit_ideal'],    'b--', lw=1.5, label='Ideal model')
ax5.axvline(opt_time,   color='red',  ls=':', lw=1.5, label=f'Hybrid: {opt_time:.1f}h')
ax5.axvline(ideal_time, color='blue', ls=':', lw=1.5, label=f'Ideal: {ideal_time:.1f}h')
# Shade the profit gain region
ax5.fill_betweenx(
    [df['profit_ensemble'].min(), df['profit_ensemble'].max()],
    opt_time, ideal_time,
    color='green', alpha=0.1, label='Time saved'
)
ax5.set_xlabel('Time (h)'); ax5.set_ylabel('Profit ($)')
ax5.set_title('Profit Optimization', fontsize=11)
ax5.legend(fontsize=8); ax5.grid(alpha=0.3)

plt.suptitle('Improved Hybrid Physics-ML Model — Industrial Reactor\n'
             '(Ensemble + Optuna tuning + Smoothing + Uncertainty quantification)',
             fontsize=13, y=1.02)
plt.savefig('/mnt/user-data/outputs/hybrid_model_improved.png', dpi=150, bbox_inches='tight')
plt.show()
print("\nPlot saved: hybrid_model_improved.png")
```
