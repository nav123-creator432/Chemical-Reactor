```python
import numpy as np
import pandas as pd
from scipy.integrate import odeint
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.model_selection import TimeSeriesSplit
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.ensemble import RandomForestRegressor
import xgboost as xgb
import lightgbm as lgb
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import optuna
import shap
optuna.logging.set_verbosity(optuna.logging.WARNING)

# ─────────────────────────────────────────────
# CONSTANTS  (Arrhenius — illustrative values)
# ─────────────────────────────────────────────
R = 8.314           # J/mol·K
Ea1, Ea2, Ea3 = 5000, 7000, 6000   # J/mol (simplified for demo)
A1,  A2,  A3  = 1.0,  0.8,  0.5   # pre-exponential factors

# ─────────────────────────────────────────────
# IDEAL MODEL  (constant T, no deactivation)
# ─────────────────────────────────────────────
def ideal_kinetics(y, t):
    A, B, C, D = y
    T = 300
    k1 = A1 * np.exp(-Ea1 / (R * T))
    k2 = A2 * np.exp(-Ea2 / (R * T))
    k3 = A3 * np.exp(-Ea3 / (R * T))
    return [-(k1+k3)*A,  k1*A - k2*B,  k2*B,  k3*A]

# ─────────────────────────────────────────────
# REAL INDUSTRIAL MODEL
# ─────────────────────────────────────────────
def real_system(y, t):
    A, B, C, D, Activity = y
    T = 300 + 10 * np.sin(2 * np.pi * t / 24)   # day-night variation
    k1 = A1 * np.exp(-Ea1 / (R * T)) * Activity
    k2 = A2 * np.exp(-Ea2 / (R * T)) * Activity
    k3 = A3 * np.exp(-Ea3 / (R * T)) * Activity
    k_deact = 0.01                                # FIXED: was 0.1 (too aggressive)
    return [-(k1+k3)*A,
             k1*A - k2*B,
             k2*B,
             k3*A,
            -k_deact * Activity * A]

# ─────────────────────────────────────────────
# GENERATE DATA
# ─────────────────────────────────────────────
np.random.seed(42)
t = np.linspace(0, 48, 1000)

ideal_sol = odeint(ideal_kinetics, [1, 0, 0, 0],    t)
real_sol  = odeint(real_system,    [1, 0, 0, 0, 1], t)

df = pd.DataFrame({
    'time':     t,
    'ideal_B':  ideal_sol[:, 1],
    'real_B':   real_sol[:,  1],
    'temp':     300 + 10 * np.sin(2 * np.pi * t / 24),
    'activity': real_sol[:, 4],
})

# ─────────────────────────────────────────────
# ADD SENSOR NOISE  (makes data more realistic)
# ─────────────────────────────────────────────
df['real_B'] += np.random.normal(0, 0.002, size=len(df))

# ─────────────────────────────────────────────
# PREPROCESSING
# ─────────────────────────────────────────────
df.loc[df.sample(frac=0.05, random_state=0).index, 'real_B'] = np.nan

imputer = SimpleImputer(strategy='median')
df['real_B_imputed'] = imputer.fit_transform(df[['real_B']])

df['B_lag1'] = df['real_B_imputed'].shift(1).bfill()   # FIX: no deprecation warning
df['B_lag2'] = df['real_B_imputed'].shift(2).bfill()   # extra lag feature
df['temp_sq'] = df['temp'] ** 2                        # non-linear temperature effect
df['residual'] = df['real_B_imputed'] - df['ideal_B']

features = ['time', 'temp', 'temp_sq', 'ideal_B', 'B_lag1', 'B_lag2', 'activity']
X = df[features].values
y = df['residual'].values

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ─────────────────────────────────────────────
# TRAIN / TEST SPLIT  (last 20% = held-out test)
# ─────────────────────────────────────────────
split = int(len(X_scaled) * 0.80)
X_train, X_test = X_scaled[:split], X_scaled[split:]
y_train, y_test = y[:split],        y[split:]

# ─────────────────────────────────────────────
# OPTUNA HYPERPARAMETER TUNING FOR XGBOOST
# ─────────────────────────────────────────────
def xgb_objective(trial):
    params = {
        'n_estimators':  trial.suggest_int('n_estimators',  100, 600),
        'max_depth':     trial.suggest_int('max_depth',      3,   8),
        'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.3, log=True),
        'subsample':     trial.suggest_float('subsample',     0.6,  1.0),
        'colsample_bytree': trial.suggest_float('colsample_bytree', 0.6, 1.0),
        'random_state':  42,
        'verbosity': 0,
    }
    tscv = TimeSeriesSplit(n_splits=5)
    maes = []
    for tr_idx, val_idx in tscv.split(X_train):
        m = xgb.XGBRegressor(**params)
        m.fit(X_train[tr_idx], y_train[tr_idx])
        maes.append(mean_absolute_error(y_train[val_idx], m.predict(X_train[val_idx])))
    return np.mean(maes)

study = optuna.create_study(direction='minimize')
study.optimize(xgb_objective, n_trials=50)
best_params = study.best_params
print("Best XGB params:", best_params)

# ─────────────────────────────────────────────
# TRAIN ALL THREE MODELS WITH BEST PARAMS
# ─────────────────────────────────────────────
xgb_model = xgb.XGBRegressor(**best_params, random_state=42, verbosity=0)
xgb_model.fit(X_train, y_train)

lgb_model = lgb.LGBMRegressor(n_estimators=400, learning_rate=0.05,
                                max_depth=5, random_state=42, verbose=-1)
lgb_model.fit(X_train, y_train)

rf_model = RandomForestRegressor(n_estimators=300, max_depth=8, random_state=42)
rf_model.fit(X_train, y_train)

# ─────────────────────────────────────────────
# HYBRID PREDICTIONS  (physics + ML residual)
# ─────────────────────────────────────────────
ideal_test = df['ideal_B'].values[split:]
real_test  = df['real_B_imputed'].values[split:]

df['hybrid_xgb'] = df['ideal_B'] + xgb_model.predict(X_scaled)
df['hybrid_lgb'] = df['ideal_B'] + lgb_model.predict(X_scaled)
df['hybrid_rf']  = df['ideal_B'] + rf_model.predict(X_scaled)

hybrid_xgb_test = ideal_test + xgb_model.predict(X_test)
hybrid_lgb_test = ideal_test + lgb_model.predict(X_test)
hybrid_rf_test  = ideal_test + rf_model.predict(X_test)

# ─────────────────────────────────────────────
# EVALUATION  (on held-out test set only)
# ─────────────────────────────────────────────
def evaluate(name, y_true, y_pred):
    mae  = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    r2   = r2_score(y_true, y_pred)
    print(f"{name:20s} | MAE={mae:.5f}  RMSE={rmse:.5f}  R²={r2:.5f}")
    return mae, rmse, r2

print("\n--- TEST SET METRICS ---")
evaluate("XGBoost Hybrid",  real_test, hybrid_xgb_test)
evaluate("LightGBM Hybrid", real_test, hybrid_lgb_test)
evaluate("RandomForest Hybrid", real_test, hybrid_rf_test)
evaluate("Ideal Model Only", real_test, ideal_test)   # baseline

# ─────────────────────────────────────────────
# FEATURE IMPORTANCE  (SHAP)
# ─────────────────────────────────────────────
explainer   = shap.Explainer(xgb_model)
shap_values = explainer(X_test)

# ─────────────────────────────────────────────
# PROFIT OPTIMIZATION
# ─────────────────────────────────────────────
def profit(yield_b, t_arr, price=1000, op_cost=20, catalyst_cost=50):
    return (yield_b * price) - (t_arr * op_cost) - catalyst_cost

df['profit_hybrid'] = profit(df['hybrid_xgb'], df['time'])
df['profit_ideal']  = profit(df['ideal_B'],    df['time'])

opt_time   = df.loc[df['profit_hybrid'].idxmax(), 'time']
ideal_time = df.loc[df['profit_ideal'].idxmax(),  'time']

print("\n--- DECISION REPORT ---")
print(f"Ideal model stop time : {ideal_time:.2f} h")
print(f"Hybrid model stop time: {opt_time:.2f} h")
print(f"Improvement           : {abs(opt_time - ideal_time):.2f} h earlier")

# ─────────────────────────────────────────────
# PLOTS
# ─────────────────────────────────────────────
fig = plt.figure(figsize=(16, 14))
gs  = gridspec.GridSpec(3, 2, figure=fig, hspace=0.42, wspace=0.32)

# 1. Concentration profiles
ax1 = fig.add_subplot(gs[0, :])
ax1.plot(t, df['real_B_imputed'], 'k-',  lw=1.5, label='Real (industrial)')
ax1.plot(t, df['ideal_B'],        'b--', lw=1.5, label='Ideal model')
ax1.plot(t, df['hybrid_xgb'],     'r-',  lw=1.5, label='Hybrid (XGBoost)')
ax1.set_xlabel('Time (h)'); ax1.set_ylabel('Conc. of B')
ax1.set_title('Concentration of product B over time')
ax1.legend(); ax1.grid(alpha=0.3)

# 2. Parity plot
ax2 = fig.add_subplot(gs[1, 0])
ax2.scatter(real_test, hybrid_xgb_test, alpha=0.4, s=8, color='steelblue')
lim = [min(real_test.min(), hybrid_xgb_test.min()),
       max(real_test.max(), hybrid_xgb_test.max())]
ax2.plot(lim, lim, 'r--', lw=1)
ax2.set_xlabel('Actual'); ax2.set_ylabel('Predicted')
ax2.set_title('Parity plot (test set)')
ax2.grid(alpha=0.3)

# 3. Residual vs time
ax3 = fig.add_subplot(gs[1, 1])
test_times = df['time'].values[split:]
ax3.scatter(test_times, real_test - hybrid_xgb_test, alpha=0.4, s=8, color='coral')
ax3.axhline(0, color='black', lw=1)
ax3.set_xlabel('Time (h)'); ax3.set_ylabel('Residual')
ax3.set_title('Residuals vs time (test set)')
ax3.grid(alpha=0.3)

# 4. SHAP summary
ax4 = fig.add_subplot(gs[2, 0])
shap.plots.bar(shap_values, max_display=7, show=False, ax=ax4)
ax4.set_title('SHAP feature importance')

# 5. Profit curves
ax5 = fig.add_subplot(gs[2, 1])
ax5.plot(t, df['profit_hybrid'], 'r-',  lw=1.5, label='Hybrid model')
ax5.plot(t, df['profit_ideal'],  'b--', lw=1.5, label='Ideal model')
ax5.axvline(opt_time,   color='red',  ls=':', lw=1)
ax5.axvline(ideal_time, color='blue', ls=':', lw=1)
ax5.set_xlabel('Time (h)'); ax5.set_ylabel('Profit')
ax5.set_title('Profit optimization')
ax5.legend(); ax5.grid(alpha=0.3)

plt.suptitle('Hybrid Physics-ML Model — Industrial Reactor', fontsize=14, y=1.01)
plt.savefig('hybrid_model_results.png', dpi=150, bbox_inches='tight')
plt.show()
print("\nPlot saved: hybrid_model_results.png")
```
